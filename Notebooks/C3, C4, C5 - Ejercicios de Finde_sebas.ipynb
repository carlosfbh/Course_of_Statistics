## Variable aleatoria discreta

El número de pinchazos medio en los neumáticos de cierto vehículo
industrial es de 0.3 por cada 50.000 km. Si el vehículo recorre 100.000 km, se pide:
1. probabilidad de que no tenga pinchazos,
2. probabilidad de que tenga menos de tres pinchazos,

Cual es el número de km. recorridos para que la probabilidad de que no tenga ningún
pinchazo sea 0.4066.


```python
# Reconocer la distribucion - REonociendo la variable
%pylab inline
```

```python
from scipy import statss
```

```python
mi_poisson=stats.poisson(0.36)
```

```python
# P(x=0)
mi_poisson.pmf(0)
```

```python
# P(X<=3)
mi_poisson.cdf(3)
```

```python
# P(X<3)
mi_poisson.cdf(2)
```

```python
for kl in range(1,200):
    k=10000*kl
    print(k)
    l=0.3*k/50000
    print(l)
    print(stats.poisson(l).pmf(0))
```

Supongamos que el consumo familiar de un cierto producto se distribuye
como una variable aleatoria de distribución uniforme, con esperanza igual a 10 y
varianza igual a 1. Determine la probabilidad de que dicho consumo esté comprendido
entre 8 y 12 unidades.

```python
# E = (a+b)/2
# Var = (b-a)**2/12
```

```python
from sympy import *
```

```python
a=Symbol("a")
b=Symbol("b")
```

```python
solve([(a+b)/2-10,((b-a)**2)/12-1])
```

Supongamos que el consumo familiar de un cierto producto se distribuye
como una variable aleatoria de distribución uniforme con a = 7 y b =25. Determine la probabilidad de que dicho consumo esté comprendido
entre 8 y 12 unidades.

```python
from scipy import stats
```

```python
mi_uniform = stats.uniform(loc=7,scale=25-7)
```

```python
xt=np.linspace(0,30,1000)
d=list(map(mi_uniform.pdf,xt))
plt.plot(xt,d)

xp=np.linspace(8,12,100)
dp=list(map(mi_uniform.pdf,xp))
plt.fill_between(x=xp,y1=dp,facecolor="red")

plt.show()
```

```python
mi_uniform.cdf(12)-mi_uniform.cdf(8)
```
