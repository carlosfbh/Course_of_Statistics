```python
%pylab inline
from scipy import stats
```

```python
# Let's create a continuous variable
mi_cont = stats.uniform(loc=20,scale=10)
```

```python
# Let's draw the DENSITY Function 
ne=np.linspace(-10,50,1000)
pe=list(map(mi_cont.pdf,ne))
plt.figure(figsize=(20,5))
plt.plot(ne,pe)
plt.show()
```

```python
# We said that the probability of x between 25 and 28 is the AREA below the cdf function
ne=np.linspace(-10,50,1000)
pe=list(map(mi_cont.pdf,ne))

a=25
b=28
ncolor=np.linspace(a,b,1000)
pcolor=list(map(mi_cont.pdf,ncolor))

plt.figure(figsize=(20,5))
plt.plot(ne,pe)
plt.fill_between(ncolor,0,pcolor,facecolor="red")
plt.show()
```

```python
# El cdf indica la probabilidad de que X sea menor de un cierto numero, es decir para un X = 28, el cdf es el area en rojo
# We said that the probability of x between 25 and 28 is the AREA below the cdf function
ne=np.linspace(-10,50,1000)
pe=list(map(mi_cont.pdf,ne))

b=28
ncolor=np.linspace(-10,b,1000)
pcolor=list(map(mi_cont.pdf,ncolor))

plt.figure(figsize=(20,5))
plt.plot(ne,pe)
plt.fill_between(ncolor,0,pcolor,facecolor="red")
plt.show()
```

```python
from ipywidgets import interact
```

```python
# Veamos como evoluciona el cdf con el valor de x. Vamos a visualizar el area 
# debajo de la funcion de densidad y el valor de ese area en la grafica abajo
@interact(x=20)
def show(x):
    ne=np.linspace(-10,50,100)
    pe=list(map(mi_cont.pdf,ne))
    ncolor=np.linspace(-10,x,100)
    pcolor=list(map(mi_cont.pdf,ncolor))

    plt.figure(figsize=(20,5))
    ax1 = plt.subplot(1,2,1)
    ax1.plot(ne,pe)
    ax1.fill_between(ncolor,0,pcolor,facecolor="red")
    
    nec=np.linspace(-10,x,100)
    pec=list(map(mi_cont.cdf,nec))
    
    ax2 = plt.subplot(1,2,2)
    ax2.plot(nec,pec)
    
    plt.show()
```
